## Aufgabe 1: 

Vergleichen Sie für einen beliebigen Namen die Geburtenrate in den US-Staaten Texas, Florida, Washington, Nebraska und Alaska! Verwenden Sie hierfür einen Plot.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [19, 10]

name = 'Daniel'
state1 = 'TX'
state2 = 'FL'
state3 = 'WA'
state4 = 'NE'
state5 = 'AK'

x1 = []
x2 = []
x3 = []
x4 = []
x5 = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []

with open("../../python_vorkurs_data/StateNames.csv", "r") as file:
    for line in file:
        data = line.strip().split(",")
    
        if data[4] == state1 and data[1] == name and data[3] == 'M':
            x1.append(int(data[2]))
            y1.append(int(data[5]))

        if data[4] == state2 and data[1] == name and data[3] == 'M':
            x2.append(int(data[2]))
            y2.append(int(data[5]))

        if data[4] == state3 and data[1] == name and data[3] == 'M':
            x3.append(int(data[2]))
            y3.append(int(data[5]))

        if data[4] == state4 and data[1] == name and data[3] == 'M':
            x4.append(int(data[2]))
            y4.append(int(data[5]))

        if data[4] == state5 and data[1] == name and data[3] == 'M':
            x5.append(int(data[2]))
            y5.append(int(data[5]))
```

```python
plt.plot(x1, y1, "-", label=state1)
plt.plot(x2, y2, "-", label=state2)
plt.plot(x3, y3, "-", label=state3)
plt.plot(x4, y4, "-", label=state4)
plt.plot(x5, y5, "-", label=state5)


plt.legend(loc='upper right')
plt.show()
```

## Aufgabe 2: 

Suchen Sie für jeden Bundesstaat das Jahr mit den meisten Auftreten des Namens und geben Sie die Ergebnisse textuell aus.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [19, 10]

name = 'Arnold'
statenames = {}

with open("../../python_vorkurs_data/StateNames.csv", "r") as file:
    for line in file:
        data = line.strip().split(",")
        
        if data[1] == name and data[3] == 'M':
            if statenames.get(data[4]) is not None:
                if int(data[5]) > statenames.get(data[4]).get('amount'):
                    statenames[data[4]] = {'year': int(data[2]), 'amount': int(data[5])}
            else:
                statenames[data[4]] = {'year': int(data[2]), 'amount': int(data[5])}
                
for key, value in statenames.items():
    print(key + ": " + str(value.get('year')) + '; ' + str(value.get('amount')))
```

## Aufgabe 3: 

Replizieren Sie den "Arnold"-Plot für die kompllette USA. Summieren Sie dafür die Geburtenzahlen aller Bundesstaaten pro Jahr auf.

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [19, 10]

name = 'Arnold'

x = []
y = []

arnold_dict = {}

with open("../../python_vorkurs_data/StateNames.csv", "r") as file:
    for line in file:
        data = line.strip().split(",")
        
        if data[1] == name and data[3] == 'M':
            if arnold_dict.get(int(data[2])) is not None:
                arnold_dict[int(data[2])] = arnold_dict.get(int(data[2])) + int(data[5])
            else:
                arnold_dict[int(data[2])] = int(data[5])
                
for key in sorted(arnold_dict.keys()):
    x.append(key)
    y.append(arnold_dict.get(key))
```

```python
plt.axvline(1984, color="black", label="Terminator 1")
plt.axvline(1991, color="orange", label="Terminator 2")
plt.axvline(2003, color="red", label="Terminator 3 / Gouverneur CA")
plt.plot(x, y)
extraticks = [1984, 1991, 2003, 2011]
plt.xticks(list(plt.xticks()[0]) + extraticks)
plt.legend(loc='upper right')
plt.show()
```
