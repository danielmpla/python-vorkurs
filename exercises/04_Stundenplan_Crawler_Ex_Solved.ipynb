```python
import requests
from bs4 import BeautifulSoup
```

Geben Sie in der folgenden Zeile die URL für ihren Stundenplan an.

```python
site = requests.get('https://service.hszg.de/stundenplan/index.php?fsel=1&sel=Stundenplan&id=&kalview=&booking=&ActiveCollege=Wintersemester+2020%2F21&SemWeek=0&ViewType=0&StudentSets=IIb20')
```

In der folgenden Zeile wird die Seite mit html5lib geladen. Der `html.parser` funktioniert in diesem Fall nicht, da das HTML-Dokument nicht valide ist.

```python
doc = BeautifulSoup(site.text, "html5lib")
```

```python
timetable = doc.find_all(attrs={'bgcolor': '#ffffff'})[0].contents
```

Lassen Sie sich die Liste ausgeben, die wir oben mit `.contents` erstellt haben.

```python
print(timetable)
```

Lassen Sie sich im folgenden das Matrikel des Stundenplans ausgeben.

```python
timetable[3]
```

```python
timetable[3].find_all(attrs={'class': 'txt2'})[1].get_text().strip()
```

Lassen Sie sich als nächstes das Datum ausgeben.

```python
timetable[3].find_all(attrs={'class':'txt'})[0].get_text().strip()
```

Als nächstes wollen wir für den ersten Tag im Stundenplan den Wochentag herausbekommen.

```python
timetable[7]
```

```python
timetable[7].find_all('img')[0]['alt']
```

Geben Sie nun für alle Stunden am ersten Tag die Art der Stunde (z.B. Vorlesung), die Zeit, die Kalenderwochen, das Modul, den Dozent und den Raum aus.

```python
for child in timetable[7].contents[1].contents[0].contents[7].find_all(attrs={'cellpadding': 3}):
    print(child.contents[1].contents[2].contents[5].get_text().strip())
    print(child.contents[1].contents[0].contents[0].get_text().strip())
    print(child.contents[1].contents[4].contents[1].get_text().strip())
    print(child.contents[1].contents[4].contents[3].get_text().strip())
    print(child.contents[1].contents[4].contents[5].get_text().strip())
    print(child.contents[1].contents[4].contents[7].get_text().strip())
    print('--- --- ---')
```

Als nächstes möchten wir uns den aktuellen Stundenplan ausgeben lassen. Suchen Sie dafür den richtigen Link von der Seite des Studentenwerks: https://www.studentenwerk-dresden.de/feeds/

```python
mensa_site = requests.get('https://www.studentenwerk-dresden.de/feeds/speiseplan.rss?mid=1')
```

```python
# mensa_site.encoding = 'utf-8'

mensa_doc = BeautifulSoup(mensa_site.text, 'xml')
```

Lassen Sie sich alle Gerichte mit Link uns Mensa-Ort (Author) ausgeben.

```python
items = mensa_doc.find_all('item')

for item in items:
    print(item.description.get_text().strip())
    print(item.link.get_text().strip())
    print(item.author.get_text().strip())
    print('--- --- ---')
```
