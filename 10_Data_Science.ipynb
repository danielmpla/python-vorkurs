```python
%matplotlib inline

import matplotlib.pyplot as plt

x = [x for x in range(11)]
y = [y ** 2 for y in range(11)]

plt.plot(x, y)
plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(11)
y = x ** 2

print('x:', type(x))

plt.plot(x, y)
plt.show()
```

### Grafiken zeichnen

```python
filtered_year = df[df['Year'] == 1960]
filtered_year_name = filtered_year[filtered_year['Name'] == 'Aaron']
filtered_year_name_sorted = filtered_year_name.sort_values('Count', ascending=False)
filtered_year_name_sorted.head()
```

```python
first_10 = filtered_year_name_sorted.iloc[:10]
first_10
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.bar(first_10['State'], first_10['Count'])
plt.show()
```
